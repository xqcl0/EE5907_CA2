# Calculate 3 covariance matrices

```python
# calculate the mean value
cmu_means = np.mean(cmu_img, axis=0)

# total covariance
data_len = len(cmu_img)
X_img = np.zeros((1024, data_len))

St = np.zeros((1024, 1024))

# convert list to array d x n where d = 1024 and n is the number of data points

for i in range(data_len):
    cmu_img[i] = cmu_img[i] - cmu_means
    cmu_img[i] = np.reshape(cmu_img[i], (1024, 1))
    X_img[:, i] = cmu_img[i][:, 0]

cmu_means = np.reshape(cmu_means, (1024, 1))
# calculate the total covariance matrix
tem = np.subtract(X_img, cmu_means)
St = tem@tem.T / data_len
# print(St)


Sb = np.zeros((1024, 1024))
class_mean = np.zeros((1024, 26))

# calculate the mean value of each class (for CMU data set)
for i in range(25):
    tem = np.mean(X_img[:, i*each_test_number:(i+1)*each_test_number], axis=1)
    class_mean[:, i] = tem

# calculate the mean value of own photos
tem = np.mean(X_img[:, each_test_number *
              sample:sample*each_test_number+6], axis=1)
class_mean[:, 25] = tem

# construct the probability term (prior)
p0 = each_test_number / data_len
p_own = 7 / data_len
p_i = [p0 for i in range(25)]
p_i.append(p_own)
p_matrix = np.diag(p_i)

# between-class covariance
tem3 = np.subtract(class_mean, cmu_means)
Sb = tem3@p_matrix@tem3.T
# print(Sb)

# Within-class covariance
Sw = St-Sb
# print(Sw)
```

# Eigenvalues and eigenvectors of S<sup>-1</sup><sub>w</sub>S<sub>b</sub>

```python
eigval, eigvec = np.linalg.eig(np.linalg.inv(Sw)@Sb)
order = np.argsort(-eigval.real)  # sort it in decrease order
eigval = eigval[order].real
eigvec = eigvec[:, order].real  # filter the imaginary part
```

# Reduce the dimentionality (projection)

```python
D = 200
cmu_reduced = np.zeros((D, data_len))
projection = np.dot(eigvec[:, :D].T, X_img)  # projection
# d by n matrix where d is dimentionality and n is number of data
cmu_reduced = projection
print(projection.shape)
```

# Processing the testing data and do dimentionality reduction

```python
test_data_len = len(cmu_test)
test_mean = np.mean(cmu_test, axis=0)
# matrix form in shape d x n where d = 1024
cmu_test_ = np.zeros((1024, test_data_len))

# convert list form to matrix from
for i in range(test_data_len):
    cmu_test[i] = cmu_test[i] - test_mean
    cmu_test[i] = np.reshape(cmu_test[i], (1024, 1))
    cmu_test_[:, i] = cmu_test[i][:, 0]

# print(cmu_test_centred.shape)
# print(cmu_train_label)
# print(cmu_train_label.shape)

# dimentionality reduction
test_reduced = np.zeros((D, data_len))
projection = np.dot(eigvec[:, :D].T, cmu_test_)
test_reduced = projection
# matrix form in shape d x n where d is the reduced dimentionality
print(test_reduced.shape)
```

# Apply K-nearest neighbors with K=10 for CMU PIE testing 

```python
for d in [2, 3, 9, 20, 100]:
    cmu_reduced = np.dot(eigvec[:, :d].T, X_img)
    test_reduced = np.dot(eigvec[:, :d].T, cmu_test_)
    estimate_label = []
    k = 1
    for i in range(test_data_len-3):
        dist = []
        for j in range(data_len):
            dist.append(np.linalg.norm(test_reduced[:d, i]-cmu_reduced[:d, j]))
        index = cmu_train_label[np.argsort(dist)[:k]]

        estimate_label.append(np.argmax(np.bincount(index)))

    count = 0
    for i in range(test_data_len-3):
        if estimate_label[i] != cmu_test_true_label[i]:
            count = count+1
    # print(count)
    accuracy = (test_data_len-count)/test_data_len
    print('accuracy using LDA by reducing dimentionality to '+str(d) +
          ' and KNN with k = '+str(k)+' is: '+str(round(100*accuracy, 2))+'%')
```

```python
import cv2
import matplotlib.pyplot as plt
for i in range(1,11):
    image = cv2.imread('./'+str(i)+'.jpg', cv2.IMREAD_GRAYSCALE)
    image = cv2.resize(image,(32,32),interpolation=cv2.INTER_CUBIC)
    plt.subplot(2,5,i)
    plt.imshow(image,cmap='gray')
    plt.axis("off")
    cv2.imwrite('./'+str(i)+'.jpg', image)
plt.show()
```
